```python
import matplotlib.pyplot as plt
import random
import time

plt.ion()  # interactive mode

x_data = []
y_data = []

for i in range(50):
    x_data.append(i)
    y_data.append(random.randint(0, 100))

    plt.clf()
    plt.plot(x_data, y_data, color='green')
    plt.title("Real-Time Smoothie Chart")
    plt.xlabel("Time")
    plt.ylabel("Value")

    plt.pause(0.5)

plt.ioff()
plt.show()
```
